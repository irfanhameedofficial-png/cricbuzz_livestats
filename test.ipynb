```python
for type_match in data['typeMatches']:
    print("Match Type:", type_match['matchType'])

    for series in type_match['seriesMatches']:
        # Some entries are ads, skip them
        if "seriesAdWrapper" not in series:
            continue  

        series_data = series['seriesAdWrapper']

        for match in series_data['matches']:
            info = match['matchInfo']
            score = match.get('matchScore', {})  # safely get matchScore

            print("Series:", info.get('seriesName'))
            print("Match:", info.get('matchDesc'))
            print("State:", info.get('state'))
            print("Status:", info.get('status'))
            print("Team 1:", info['team1']['teamName'])
            print("Team 2:", info['team2']['teamName'])
            print("Venue:", info['venueInfo']['ground'], "-", info['venueInfo']['city'])

            # Team 1 score (if available)
            if 'team1Score' in score:
                inngs1 = score['team1Score'].get('inngs1', {})
                print("Team 1 Score:", 
                      inngs1.get('runs', '-'), "/", 
                      inngs1.get('wickets', '-'), 
                      "in", inngs1.get('overs', '-'), "overs")
            else:
                print("Team 1 Score: Not available")

            # Team 2 score (if available)
            if 'team2Score' in score:
                inngs1 = score['team2Score'].get('inngs1', {})
                print("Team 2 Score:", 
                      inngs1.get('runs', '-'), "/", 
                      inngs1.get('wickets', '-'), 
                      "in", inngs1.get('overs', '-'), "overs")
            else:
                print("Team 2 Score: Not available")

            print("-" * 40)
```

```python
#Q1,Players who represented India 
import requests

url = "https://cricbuzz-cricket.p.rapidapi.com/teams/v1/2/players"

headers = {
	"x-rapidapi-key": "fe9c1d0ff9mshf8e519597a3627fp1b6d45jsn79f69a8379bc",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

print(response.json())
```

```python
data = response.json()
data['player']
for player in data['player']:
    if player.get('name')=='BATSMEN':
        role = "BATSMEN"
        continue
    if player.get('name')=='ALL ROUNDER':
        role = "ALL ROUNDER"
        continue
    if player.get('name')=='WICKET KEEPER':
        role = "WICKET KEEPER"
        continue
    if player.get('name')=='BOWLER':
        role = "BOWLER"
        continue

    print(player.get('id'),player.get('name'),role,player.get('battingStyle'),player.get('bowlingStyle'))
    cursor.execute("""insert into indian_players values(%s,%s,%s,%s,%s) """,(player.get('id'),player.get('name'),role,player.get('battingStyle'),player.get('bowlingStyle')))
    conn.commit()
```
